# STEP -04 SPLIT DATA SET INTO INDEPENDENT/DEPENDENT VARAIBLE

```python
target=df['Insulin']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['Glucose']], target, test_size=0.2, random_state=4)
```

# STEP-05 TRAIN REGRESSION MODEL
# STEP -06 PREDICT THE RESULT 
# STEP-07 PLOT THE TRAINING AND TEST RESULTS 

```python
from sklearn.linear_model import LinearRegression

# Instantiate the model
model = LinearRegression()

# Fit the model on training data
model.fit(X_train[['Glucose']], y_train)  # Assuming 'Glucose' is the feature used for training

# Predict on training set
y_pred_train = model.predict(X_train[['Glucose']])

# Predict on test set (if needed for another plot or evaluation)
y_pred_test = model.predict(X_test[['Glucose']])

# Now plot
plt.scatter(X_test['Glucose'], y_test, color='lightcoral', label='Test Set')
plt.plot(X_train['Glucose'], y_pred_train, color='firebrick', label='Model Prediction')
plt.title('Glucose vs Insulin (Test Set)')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.legend(title='Legend', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
